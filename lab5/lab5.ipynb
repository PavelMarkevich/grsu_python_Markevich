При решении научных задач с помощью численных методов зачастую приходится иметь дело с большими массивами данных в виде массивов и матриц. Выберите тригонометрическую функцию $T(x)$, выберите полином $P(x)$. Постройте график функции $T(x)P(x)$.

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.sin(x) - 2
P = 2*x + 4*x*x
fig, ax = plt.subplots()
ax.plot(x, T)
ax.plot(x, P)
plt.show()
```

К графику добавьте: обозначение оси абсцисс, обозначение оси ординат, название графика, вставка легенды (текста в label).

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.sin(x) - 2
P = 2*x + 4*x*x
fig, ax = plt.subplots()
ax.plot(x, T, label = "sin(x)")
ax.plot(x, P, label = "2*x + 4*x*x")
plt.title("T(x)P(x)") 
plt.xlabel("x")
plt.ylabel("y") 
ax.legend()
```

Измените цвет графика на Ваш любимый. 

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.sin(x) - 2
P = 2*x + 4*x*x
fig, ax = plt.subplots()
ax.plot(x, T, label = "sin(x)", color = "k")
ax.plot(x, P, label = "2*x + 4*x*x", color = "g")
plt.title("T(x)P(x)") 
plt.xlabel("x")
plt.ylabel("y") 
ax.legend()
```

Постройте несколько квивых. Первая - цвет красный, сплошная, вторая - цвет синий, штрих-пунктир.

Измените размер на 300x300

```python
import cv2
image = cv2.imread('image.jpg',1)
img = cv2.resize(image, (300, 300))
cv2.imshow("3", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Разбейте ее на 2 картинки. Сохраните. 

```python
import cv2
import numpy as np

img = cv2.imread('image.jpg',1)
(h,w,c) = img.shape
parts = []
step_x = 2
step_y = 1
for x in range(step_x):
    for y in range(step_y):
        xratio1 = x/step_x
        xratio2 = (x+1)/step_x
        yratio1 = y/step_y
        yratio2 = (y+1)/step_y
        part = img[int(yratio1*h):int(yratio2*h),int(xratio1*w):int(xratio2*w)].copy()
        parts.append(part)
cv2.imwrite("flip1.png", parts[0])
cv2.imwrite("flip2.png", parts[1])
cv2.waitKey(0)
```

Прочитайте 2 картинки и склейте. 

```python
import cv2
img1 = cv2.imread('flip1.png',1)
img2 = cv2.imread('flip2.png',1)
vis = np.concatenate((img1, img2), axis=1) 
cv2.imshow("4", vis)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
